```python
from tkinter import *
import tkinter as tk
from PIL import ImageTk,Image

OPTIONS = [
    "California",
    "Nevada",
    "Texas"
]

#Creates the instance
window = tk.Tk()

#Titles
window.title("Corona")
window.geometry("800x500")

#drop down boxes
clicked = StringVar()
clicked.set(OPTIONS[0])

drop = OptionMenu(window, clicked, *OPTIONS)
drop.config(width=20, font=('Helvetica', 12))
drop.pack()

window.mainloop()
```
